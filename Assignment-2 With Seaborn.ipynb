## 2: Plot 12 most dangerous cities

```python
kill.city.value_counts().head(13)
```

```python
kill.city.value_counts().head(13).unique()
```

```python
top_12_dangerous_cities = kill.city.value_counts().head(12)
```

```python
top_12_dangerous_cities.index
```

```python
top_12_cities = kill["city"].value_counts().head(12)

fig, ax = plt.subplots(figsize=(9,5))

ax = sns.barplot(x = top_12_cities.index,
            y = top_12_cities.values, data = kill, ax=ax)

plt.xticks(rotation = 45, color = "blue");
ax.set_title("Most dangerous cities", color = "red")
plt.show()
```

```python
# with seaborn
```

```python
fig, ax = plt.subplots(figsize=(6,4))

ax = sns.countplot(x = "city", data = kill, order = top_12_cities.index, ax=ax)

plt.xticks(rotation = 45, size=8);
ax.set_title("Most dangerous cities", color = "blue")
plt.show()
```

```python
# with seaborn countplot
```
